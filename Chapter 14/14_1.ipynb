# Staking Multiple feature maps

```python
from sklearn.datasets import load_sample_image
import numpy as np
import matplotlib.pyplot as plt
```

```python
chaina = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") / 255
images = np.array([chaina, flower])
batch_size, height, weidth, channels = images.shape
```

```python
batch_size, height, weidth, channels
```

```python
# create two filters
filters = np.zeros(shape=(7, 7, channels, 2), dtype =np.float32)
filters[:, 3, :, 0] = 1 # vertical line
filters[3, :, :, 1] = 1 # horizontal line
```

```python
filters[3, :, :, 1]
```

```python
import tensorflow as tf
outputs = tf.nn.conv2d(images, filters, strides=1, padding="SAME")
```

```python
plt.imshow(outputs[0, :, :, 1], cmap="gray") # plot 1st image's 2nd feature map
plt.show()
```

```python
plt.imshow(outputs[1, :, :, 1], cmap="gray") # plot 2nd image
```
